```python
print("my name is {0} , my age is {1}".format("Ahmed", 20))

print("my name is {} , my age is {}".format("Ahmed", 20))

print("my name is {x} , my age is {y}".format(x="Ahmed", y=20))
```

```python
name = "Ahmed"
age = 20
print(f"my name is {name}, my age is {age}")
```

```python
input_list = ["Ahmed", "Salah", "python", "machine learning", "data science", "deep learning"]

item_to_remove = input("Enter the item you want to remove from the list: ")

if item_to_remove in input_list:
    input_list.remove(item_to_remove)
    print("\n Item removed successfully!")

print("The list after Delete:")
print(input_list)
```

```python
num1_str = input("Enter the first number: ")

num2_str = input("Enter the second number: ")

try:
    num1 = float(num1_str)
    num2 = float(num2_str)
except ValueError:
    print("Invalid input. Please enter valid numbers.")
    # Stop the program if the conversion fails
    exit()



addition_result = num1 + num2

subtraction_result = num1 - num2

multiplication_result = num1 * num2

divide_result= num1/num2 


print("\n--- Results ---")
print(f"Addition ({num1} + {num2}): {addition_result}")
print(f"Subtraction ({num1} - {num2}): {subtraction_result}")
print(f"Multiplication ({num1} * {num2}): {multiplication_result}")
print(f"Division ({num1} / {num2}): {division_result}")
```

```python
degree = float(input("Enter the student's degree: "))

if 90 <= degree <= 100:
    grade = "A"
elif 80 <= degree < 90:
    grade = "B"
elif 70 <= degree < 80:
    grade = "C"
elif 60 <= degree < 70:
    grade = "D"
else:
    grade = "F"

print("The student's grade is:", grade)
```

```python
def automated_stat_analyzer(df, column_name):
    s = df[column_name].dropna()
    if pd.api.types.is_numeric_dtype(s):
        mean = s.mean()
        median = s.median()
        std = s.std()
        skewed = "Skewed" if mean != median else "Not Skewed"
        return {"Mean": mean, "Median": median, "Standard Deviation": std, "Skewness": skewed}
    else:
        m = s.mode()
        return {"Mode": m.iloc[0] if not m.empty else None}
```

```python
# Another Way
import pandas as pd
import numpy as np

def automated_stat_analyzer(df, column_name):
    # 1. Identify the column data
    col = df[column_name]
    
    # Bonus: Check if the column is categorical (non-numeric)
    if not pd.api.types.is_numeric_dtype(col):
        # Calculate Mode (the first one in case of multiple)
        mode_val = col.mode()[0]
        return {
            "Column": column_name,
            "Type": "Categorical",
            "Mode": mode_val
        }

    # 2. Calculate Mean, Median, and Standard Deviation
    # Pandas ignores NaNs by default in these calculations
    mean_val = col.mean()
    median_val = col.median()
    std_val = col.std()

    # 3. Identify Skewness
    # Positively Skewed: Mean > Median
    # Negatively Skewed: Mean < Median
    # Symmetric: Mean == Median
    if mean_val > median_val:
        skew_type = "Positively Skewed (Right)"
    elif mean_val < median_val:
        skew_type = "Negatively Skewed (Left)"
    else:
        skew_type = "Symmetric"

    # Return results in a structured format
    return {
        "Column": column_name,
        "Type": "Numerical",
        "Mean": round(mean_val, 2),
        "Median": round(median_val, 2),
        "Standard Deviation": round(std_val, 2),
        "Skewness": skew_type
    }


sales_stats = automated_stat_analyzer(df_test, 'Sales_Amount')
print(f"Sales Stats: {sales_stats}")

category_stats = automated_stat_analyzer(df_test, 'Product_Category')
print(f"Category Stats: {category_stats}")

age_stats = automated_stat_analyzer(df_test, 'Customer_Age')
print(f"Age Stats: {age_stats}")
```

```python
import pandas as pd

def null_handling_strategy(df, strategy="fill_mean"):
    """
    Cleans a dataset by resolving missing (NaN) values based on a chosen strategy.
    """
    # Create a copy to avoid modifying the original dataframe
    df_cleaned = df.copy()
    
    # Check if there are any nulls at all
    if not df_cleaned.isnull().values.any():
        print("No null values detected. Returning original data.")
        return df_cleaned

    if strategy == "drop_rows":
        # Removes any row containing at least one null value
        df_cleaned = df_cleaned.dropna()
        
    elif strategy in ["fill_mean", "fill_median"]:
        # Identify numerical columns (int and float)
        num_cols = df_cleaned.select_dtypes(include=['number']).columns
        
        for col in num_cols:
            if strategy == "fill_mean":
                fill_value = df_cleaned[col].mean()
            else:
                fill_value = df_cleaned[col].median()
                
            df_cleaned[col] = df_cleaned[col].fillna(fill_value)
            
    else:
        print(f"Strategy '{strategy}' not recognized. No changes made.")

    return df_cleaned
```

```python
def factorial (num):
    if num == 0 :
        return 0
    if num ==1:
        return 1
    return num * factorial(num-1)

factorial(5)
```

```python
def dividors(num1,num2):
    """
    param1: the function take num1 of the dividors
    type:param_1 = int
    
    """
    mn= min(num1,num2)
    common_diviadors =[]
    for i in range (1, mn+1):
        if num1 % i == 0 and num2 %i ==0:
            common_dividors.append(i)
    return common_dividors
```

```python
def fun_alternative(s):
    words = s.split()
    
    reversed_words = reversed(words) 
    
    result = ' '.join(reversed_words)
        
    return result

s = "Ahmed Salah"
result = fun_alternative(s)
print(result)
```

```python
import secrets
import string

def generate_strong_password(length=12):
    
    lower_case = string.ascii_lowercase
    upper_case = string.ascii_uppercase
    digits = string.digits
    symbols = string.punctuation
    
    
    all_characters = lower_case + upper_case + digits + symbols
    
    
    
    password = []
    password.append(secrets.choice(lower_case))
    password.append(secrets.choice(upper_case))
    password.append(secrets.choice(digits))
    password.append(secrets.choice(symbols))
    
    
    
    for _ in range(length - 4):
        password.append(secrets.choice(all_characters))
        
    
    secrets.SystemRandom().shuffle(password)
    
    
    return "".join(password)


password_12 = generate_strong_password()
print(f"Pasword length 12{password_12}")

password_18 = generate_strong_password(length=18)
print(f"Pasword length 18{password_18}")
```

```python
import random

def generate_passcode(length):
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*()"
    password = ""
    for _ in range(length):
        password += random.choice(chars)
    return password
```

```python
def count_words_frequency(word_list):
    """
    Counts the frequency of each unique word in a list using the procedure
    shown in the image (set conversion, dictionary initialization, nested loops).

    Args:
        word_list (list): A list of words (strings).

    Returns:
        dict: A dictionary where keys are unique words and values are their counts.
    """
    unique_words = set(word_list)

    counts = {word: 0 for word in unique_words}

   
    for i in unique_words:
        for j in word_list:
            if i == j:
                counts[i] += 1

    return counts

words = ["welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]

word_counts = count_words_frequency(words)

print(f"Original Words: {words}")
print(f"Word Numbers: {word_counts}")
```

```python
import random

class Task9:
    """
    Task9 class representing a simple keyword-based chatbot.
    
    Attributes:
        responses (dict): A dictionary mapping keywords to lists of potential responses.
    """

    def __init__(self, responses):
        """
        Initializes the chatbot with a set of responses.
        
        Args:
            responses (dict): The dictionary containing keywords and response lists.
        """
        self.responses = responses

    def get_response(self, user_input):
        """
        Processes the user's input and returns a matching response from the dictionary.
        
        Args:
            user_input (str): The text entered by the user.
            
        Returns:
            str: A randomly selected response if a keyword is found, otherwise a default message.
        """
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        
        return random.choice(self.responses["default"])

    def start_chat(self):
        """
        Starts the interactive chat loop, taking input from the user and printing responses.
        The loop terminates when the user says 'goodbye'.
        """
        print("chatbot: hi how can i help you?")
        
        while True:
            user_input = input("user: ").lower()
            response = self.get_response(user_input)
            print("chatbot:", response)
            
            if "goodbye" in user_input:
                break

# --- Execution ---

# 1. Define the responses dictionary
chat_data = {
    "hello": ["hello!", "Hi there!", "Greetings!"],
    "how are you": ["I'm doing well, thank you", "i'm fine, how are you?"],
    "goodbye": ["Goodbye", "see you later", "farewell!"],
    "default": ["i'm sorry, i did not understand. ", "could you please rephrase that?"]
}

# 2. Create an object from the class
my_chatbot = Task9(chat_data)

# 3. Call the main function using the object
my_chatbot.start_chat()
```

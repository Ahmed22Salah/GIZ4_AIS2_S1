# <a id="14"> 14 )  .  What is the  Correlation between `specific numeric variables` in the dataset?</a>


```python
# Make visualization to easy get insight

corr =df.corr(numeric_only=True)['age'].sort_values()


plt.figure(figsize=(10,5) ,dpi =200)
sns.barplot(x= corr.values ,y =corr.index )
plt.title(f"the Correlation of Age")
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()

```

> ### <mark>Your analsys and exploration here </mark>

Example usage
- There is `no Correlation` for any variable with time ( `duration sec /minutes`)




- There is small negative correlation between the longitude and small positive correlation with latitude.    
  There is strong negative correlation with member_birth_year




- bike_id nearly has no correlation with any feature

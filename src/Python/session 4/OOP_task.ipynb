```python
#Task1: Factorial Calculator
class FactorialTask:
    
    def __init__(self, number):
        self.number = number

    def calculate(self, num=None):
        if num is None:
            num = self.number
        if num == 0 or num == 1:
            return 1
        return num * self.calculate(num - 1)

if __name__ == "__main__":
    task1 = FactorialTask(5)
    print(f"Factorial of 5: {task1.calculate()}")
```

```python
#Task 2: Prime Number Checker
class PrimeChecker:
    def __init__(self, number):
        self.number = number

    def is_prime(self):
        if self.number < 2:
            return False
        for i in range(2, int(self.number ** 0.5) + 1):
            if self.number % i == 0:
                return False
        return True

if __name__ == "__main__":
    task2 = PrimeChecker(22)
    print(f" {task2.is_prime()}")
```

```python
#Task 3: Common Divisors Finder
class DivisorFinder:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def get_common_divisors(self):
        mn = min(self.num1, self.num2)
        common = []
        for i in range(1, mn + 1):
            if self.num1 % i == 0 and self.num2 % i == 0:
                common.append(i)
        return common

if __name__ == "__main__":
    task3 = DivisorFinder(20, 10)
    print(f"Common divisors of : {task3.get_common_divisors()}")
```

```python
#Task 4: Longest Common Subsequence
class StringMatcher:
    def __init__(self, str1, str2):
        self.str1 = str1
        self.str2 = str2

    def find_lcs(self):
        longest_str = ""
        for i in range(len(self.str1)):
            for j in range(i, len(self.str1)):
                sub_str = self.str1[i:j+1]
                if sub_str in self.str2:
                    if len(sub_str) > len(longest_str):
                        longest_str = sub_str
        return longest_str

if __name__ == "__main__":
    task4 = StringMatcher("Welcome to Machine Learning", "Studying Machine Learning Course")
    print(f"Longest Common: {task4.find_lcs()}")
```

```python
#Task 5: Word Reverser
class SentenceReverser:
    def __init__(self, text):
        self.text = text

    def reverse(self):
        words = self.text.split(" ")
        return " ".join(words[::-1])

if __name__ == "__main__":
    task5 = SentenceReverser("Hello World")
    print(f"Reversed: {task5.reverse()}")
```

```python
import random

class PasswordGenerator:
    """
    Class to generate random secure passwords.
    
    """
    
    def __init__(self, length):
        """
        Initializes the desired password length.
        
        """
        self.length = length
        self.chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"

    def generate(self):
        """
        Creates a random string of the specified length.
        
        """
        password = ""
        for _ in range(self.length):
            rand_idx = random.randrange(0, len(self.chars))
            password += self.chars[rand_idx]
        return password

if __name__ == "__main__":
    task6 = PasswordGenerator(12)
    print(f"Random Password: {task6.generate()}")
```

```python
#Task 7: Word Frequency Counter
class WordCounter:
    """
    Class to count unique word frequencies.
    """
    
    def __init__(self, word_list):
        """
        Initializes the list of words.
        """
        self.words = word_list

    def get_frequencies(self):
        """
        Returns a dictionary with word counts.
        """
        unique_words = set(self.words)
        return {word: self.words.count(word) for word in unique_words}

if __name__ == "__main__":
    words = ["Ali", "Hi", "Ali", "No", "Hi", "Ali"]
    task7 = WordCounter(words)
    print(f"Word Frequencies: {task7.get_frequencies()}")
```

```python
#Task 8: Even/Odd
class NumberFilter:
    """
    Class to filter even numbers from a range.
    """
    
    def __init__(self, limit):
        """
        Initializes the upper limit of the range.
        """
        self.limit = limit

    def get_evens(self):
        """
        Returns a list of even numbers up to the limit.
        """
        return [i for i in range(self.limit) if i % 2 == 0]

if __name__ == "__main__":
    task8 = NumberFilter(10)
    print(f"Evens up to 10: {task8.get_evens()}")
```

```python
import random

class Chatbot:
    """
    Chatbot class handles the interactive logic for responding to user keywords.
    
    Attributes:
        responses (dict): A dictionary mapping keywords to lists of potential responses.
    """

    def __init__(self):
        """
        Initializes the chatbot with the predefined dictionary of responses.
        """
        self.responses = {
            "hello": ["Hello!", "Hi there!", "Greetings!"],
            "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
            "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
            "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }

    def get_response(self, user_input):
        """
        Finds a matching keyword in the user input and selects a random response.
        
        Args:
            user_input (str): The text entered by the user.
            
        Returns:
            str: A randomly selected response from the dictionary.
        """
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        
        return random.choice(self.responses["default"])

    def start_chat(self):
        """
        Starts the chat interface loop. 
        Breaks the loop when the user enters 'goodbye'.
        """
        print("Chatbot: Hi! How can I assist you today?")
        
        while True:
            user_input = input("User: ").lower()
            
            response = self.get_response(user_input)
            print("Chatbot:", response)
            
            if user_input == "goodbye":
                break


my_bot = Chatbot()
my_bot.start_chat()
```

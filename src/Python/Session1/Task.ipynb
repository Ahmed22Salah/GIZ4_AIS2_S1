```python
#Task-1
#Suppose you have this email address “Amit_ml@gmail.edu”
email_address = "Amit_ml@gmail.edu"

print("--- Task 1: Email Processing ---")

if email_address.count('@') != 1:
    print("Validation: Invalid email (More or less than one '@').")
else:
    parts = email_address.split('@')
    username = parts[0]
    domain_part = parts[1]

    if '.' not in domain_part:
        print("Validation: Invalid email (No dot after '@').")
    else:
        print("Extract Username:", username)

        last_dot_index = domain_part.rfind('.')
        domain = domain_part[:last_dot_index]
        print("Extract Domain:", domain)
        
        domain_ending = domain_part[last_dot_index:]
        
        if domain_ending == ".com":
            print("Check for Domain Ending: Commercial Domain")
        elif domain_ending == ".edu":
            print("Check for Domain Ending: Educational Domain")
        else:
            print("Check for Domain Ending: Other Domain")
```

```python
#Task-2:
#Encoded Message: ###!!@mocleW EPGTQ!!!6789
encoded_message_2 = "###!!@mocleW EPGTQ!!!6789"

print("\n--- Task 2: Decode Message (Welcome PGTQ) ---")

core_message_2 = encoded_message_2[6:-7]

words_2 = core_message_2.split()
word1_2 = words_2[0]
word2_2 = words_2[1]

reversed_word_2 = word1_2[::-1]

modified_word2_2 = word2_2[1:]

final_decoded_message_2 = reversed_word_2 + " " + modified_word2_2

print("Final decoded message:", final_decoded_message_2)
```

```python
#Task-3:
#Encoded Message: &&&**$gnirtS PLIO!!@1234
encoded_message_3 = "&&&**$gnirts PLIO!!@1234"

print("\n--- Task 3: Decode Message (String PLEU) ---")

core_message_3 = encoded_message_3[6:-6]

words_3 = core_message_3.split()
word1_3 = words_3[0]
word2_3 = words_3[1]

reversed_word_3 = word1_3[::-1].capitalize()

modified_word2_3 = word2_3.replace('I', 'E').replace('O', 'U')

final_decoded_message_3 = reversed_word_3 + " " + modified_word2_3

print("Final decoded message:", final_decoded_message_3)
```

```python
#Task-4:
#Encoded Message: ##$$$@!yalpstcejorp EPUVT****9887
encoded_message_4 = "##\$$\$@!yalpstcejorp EPUVT****9887"

print("\n--- Task 4: Decode Message (projectsalpy APOVT) ---")

core_message_4 = encoded_message_4[8:-8]

words_4 = core_message_4.split()
word1_4 = words_4[0]
word2_4 = words_4[1]

reversed_word_4 = word1_4[::-1]

modified_word2_4 = word2_4.replace('E', 'A').replace('U', 'O')

final_decoded_message_4 = reversed_word_4 + " " + modified_word2_4

print("Final decoded message:", final_decoded_message_4)
```
